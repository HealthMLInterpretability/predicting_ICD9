```python
def correct_age(df1: pd.DataFrame) -> pd.DataFrame:
    
    #create copy 
    df = df1.copy()
    #Create look-up table of mean age based off of SUBJECT ID
    SUBJECT_ID_LOOKUP = df[df['AGE']!=0].groupby('SUBJECT_ID').mean()['AGE']
    
    #Replace values where AGE == 0. However note that some subjects only have AGE == 0 values
    for i in range(len(df)):
        if df['AGE'].iloc[i] == 0:
            try:
                df['AGE'].iloc[i] = SUBJECT_ID_LOOKUP.loc[df['SUBJECT_ID'].iloc[i]]
            except:
                continue
    return df
```
