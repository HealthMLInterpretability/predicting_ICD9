## Age Extraction Functions {-}

Below are functions to be used:

```python
import re
```

```python
# function for extracting Age as Number only from TEXT column (use after removing missing ages from entire data)

def find_age(text):
    num = (re.search(r'\d+(?=yo)',text) or re.search(r'\d+(?= yo)',text) or re.search(r'\d+(?=-yo )',text)
    or re.search(r'\d+(?=y/o)',text) or re.search(r'\d+(?= y/o)',text) 
    or re.search(r'\d+(?=Y/O)',text) or re.search(r'\d+(?= Y/O)',text) or re.search(r'\d+(?= -yo)',text)
    or re.search(r'\d+(?=-year-old)',text) or re.search(r'\d+(?= year old)',text) or re.search(r'\d+(?= years old)',text)
    or re.search(r'\d+(?=-years-old)',text) or re.search(r'\d+(?= year-old)',text) or re.search(r'\d+(?= years-old)',text)
    or re.search(r'\d+(?=yoF)',text) or re.search(r'\d+(?=yoM)',text) or re.search(r'\d+(?= -Y/O)',text)
    or re.search(r'\d+(?= yoF)',text) or re.search(r'\d+(?= yoM)',text) 
    or re.search(r'\d+(?=M )',text)
    or re.search(r'\d+(?= M )',text)
    or re.search(r'\d+(?= years/old)',text) or re.search(r'\d+(?=years/old)',text)
    or re.search(r'\d+(?=YOM)',text) or re.search(r'\d+(?=YOF)',text)
    or re.search(r'\d+(?= YOM)',text) or re.search(r'\d+(?= YOF)',text)
    or re.search(r'\d+(?=y )',text) or re.search(r'\d+(?= y )',text) or re.search(r'\d+(?= y.o)',text) or re.search(r'\d+(?=y.o)',text)
    or re.search(r'\d+(?= y.o.)',text) or re.search(r'\d+(?=y.o.)',text)
    or re.search(r'\d+(?=F )',text) or re.search(r'\d+(?= F )',text)).group()
    
    return num
```

```python
# function for extracting Age as re.match from TEXT column (use to find missing ages from entire data)

def find_match(text):
    num = (re.search(r'\d+(?=yo)',text) or re.search(r'\d+(?= yo)',text) or re.search(r'\d+(?=-yo )',text)
    or re.search(r'\d+(?=y/o)',text) or re.search(r'\d+(?= y/o)',text) 
    or re.search(r'\d+(?=Y/O)',text) or re.search(r'\d+(?= Y/O)',text) or re.search(r'\d+(?= -yo )',text)
    or re.search(r'\d+(?=-year-old)',text) or re.search(r'\d+(?= year old)',text) or re.search(r'\d+(?= years old)',text)
    or re.search(r'\d+(?=-years-old)',text) or re.search(r'\d+(?= year-old)',text) or re.search(r'\d+(?= years-old)',text)
    or re.search(r'\d+(?=yoF)',text) or re.search(r'\d+(?=yoM)',text) or re.search(r'\d+(?= -Y/O)',text)
    or re.search(r'\d+(?= yoF)',text) or re.search(r'\d+(?= yoM)',text)
    or re.search(r'\d+(?=M )',text)
    or re.search(r'\d+(?= years/old)',text) or re.search(r'\d+(?=years/old)',text)
    or re.search(r'\d+(?= M )',text)
    or re.search(r'\d+(?=YOM)',text) or re.search(r'\d+(?=YOF)',text)
    or re.search(r'\d+(?= YOM)',text) or re.search(r'\d+(?= YOF)',text)
    or re.search(r'\d+(?=y )',text) or re.search(r'\d+(?= y )',text)
    or re.search(r'\d+(?= y.o.)',text) or re.search(r'\d+(?=y.o.)',text) or re.search(r'\d+(?= y.o)',text) or re.search(r'\d+(?=y.o)',text)
    or re.search(r'\d+(?=F )',text) or re.search(r'\d+(?= F )',text))
    
    return num
```

## Finding Age Matches in Text Column {-}

```python
noteevents_df['AGE_MATCH'] = noteevents_df['TEXT'].apply(find_match)
noteevents_df
```

```python
noteevents_df['AGE_MATCH'].isnull().sum()
```

## Placing None Value Rows Into A New DataFrame {-}

```python
noteevents_df_age_none = noteevents_df[noteevents_df['AGE_MATCH'].isnull()]
noteevents_df_age_none
```

## Dropping None Values in Original DataFrame {-}

```python
noteevents_df_drop_none = noteevents_df.dropna(axis=0, subset=['AGE_MATCH'])
noteevents_df_drop_none
```

```python
noteevents_df_drop_none['AGE_MATCH'].isnull().sum()
```

```python
noteevents_df_drop_none['HADM_ID'].isnull().sum()
```

## Applying Function To Get Age Value As Number {-}

```python
# creating a new column called age and applying function on full data
noteevents_df_drop_none['AGE'] = noteevents_df_drop_none['TEXT'].apply(find_age)

noteevents_df_drop_none
```

```python
noteevents_df_drop_none.info()
```

## Concatenating Both DataFrames {-}

```python
# noteevents_df_full = noteevents_df_age_none.merge(noteevents_df, left_index=True, right_index=True)
# noteevents_df_full
```
