## Concat DataFrames {-}

```python
dia_train_df = pd.read_csv('/Users/JoKerry/clinical-outcome-prediction/ICD_prediction/DIA_GROUPS_3_DIGITS_adm_train.csv')
dia_test_df = pd.read_csv('/Users/JoKerry/clinical-outcome-prediction/ICD_prediction/DIA_GROUPS_3_DIGITS_adm_test.csv')
dia_val_df = pd.read_csv('/Users/JoKerry/clinical-outcome-prediction/ICD_prediction/DIA_GROUPS_3_DIGITS_adm_val.csv')
```

```python
dia_val_df
```
