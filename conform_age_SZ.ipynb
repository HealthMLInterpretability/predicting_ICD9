```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#load first 10000 rows for testing
df = pd.read_csv('NOTEEVENTS_WITH_AGE.csv', nrows = 10000)
```

```python
df[df['AGE']==0]['SUBJECT_ID'].nunique()
#858 unique subjects where age == 0
```

```python
def correct_age(df1: pd.DataFrame) -> pd.DataFrame:
    
    #create copy 
    df = df1.copy()
    #Create look-up table of mean age based off of SUBJECT ID
    SUBJECT_ID_LOOKUP = df[df['AGE']!=0].groupby('SUBJECT_ID').mean()['AGE']
    
    #Replace values where AGE == 0. However note that some subjects only have AGE == 0 values
    for i in range(len(df)):
        if df['AGE'].iloc[i] == 0:
            try:
                df['AGE'].iloc[i] = SUBJECT_ID_LOOKUP.loc[df['SUBJECT_ID'].iloc[i]]
            except:
                continue
    return df
```

```python
#apply function to correct age
df_corrected = correct_age(df)
```
